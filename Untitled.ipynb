```python
# ===== 2. LÀM SẠCH =====
df = df.dropna()
```

```python
# ===== 3. XỬ LÝ DATE =====
df["Date"] = pd.to_datetime(df["Date"], errors='coerce', dayfirst=True)

# Xóa dòng lỗi Date
df = df.dropna(subset=["Date"])

# Sắp xếp theo ngày
df = df.sort_values(by="Date")

# ===== 4. CHUẨN HÓA GIÁ =====
for col in df.columns:
    if col != "Date":
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].apply(lambda x: round(x/1000, 2) if x > 1000 else x)

# ===== 5. FORMAT DATE =====
df["Date"] = df["Date"].dt.strftime("%m/%d/%Y")

# ===== 6. LƯU FILE SAU XỬ LÝ =====
df.to_excel("stock_cleaned.xlsx", index=False)

print("✅ Đã làm sạch dữ liệu!")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ===== FIX FONT =====
plt.rcParams['font.family'] = 'DejaVu Sans'

# ===== ĐỌC FILE =====
df = pd.read_excel("stock_cleaned.xlsx")
df['Date'] = pd.to_datetime(df['Date'])

# ===== LỌC DATA =====
df = df[df['Date'] >= "2020-01-01"]

stocks = ["VCB", "VIC", "VHM", "BID", "HPG"]

# ===== STYLE DASHBOARD =====
plt.style.use('seaborn-v0_8-darkgrid')

plt.figure(figsize=(15,6))

colors = {
    "VCB": "#1f77b4",
    "VIC": "#e41a1c",
    "VHM": "#4daf4a",
    "BID": "#ff7f00",
    "HPG": "#984ea3"
}

# ===== VẼ =====
for stock in stocks:
    if stock in df.columns:
        smooth = df[stock].rolling(window=7).mean()

        plt.plot(df['Date'], smooth,
                 label=stock,
                 color=colors[stock],
                 linewidth=3,
                 alpha=0.9)

        # highlight điểm cuối
        plt.scatter(df['Date'].iloc[-1],
                    smooth.iloc[-1],
                    color=colors[stock],
                    s=60)

        # ghi giá cuối
        plt.text(df['Date'].iloc[-1],
                 smooth.iloc[-1],
                 f"{round(smooth.iloc[-1],1)}",
                 fontsize=9)

# ===== TRANG TRÍ =====
plt.title("Biểu đồ giá cổ phiếu 5 công ty (2020 - nay)",
          fontsize=16, weight='bold')

plt.xlabel("Ngày", fontsize=11)
plt.ylabel("Giá cổ phiếu", fontsize=11)

# grid đẹp hơn
plt.grid(alpha=0.2)

# legend dưới
plt.legend(loc='upper center',
           bbox_to_anchor=(0.5, -0.15),
           ncol=5,
           frameon=False,
           fontsize=11)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
import ipywidgets as widgets
from IPython.display import display

# ===== LOAD DATA =====
df = pd.read_excel("stock_cleaned.xlsx")
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date'] >= "2020-01-01"]

stocks = ["VCB", "VIC", "VHM", "BID", "HPG"]

# ===== DROPDOWN =====
dropdown = widgets.Dropdown(
    options=stocks,
    value='VCB',
    description='Chọn mã:',
)

date_input = widgets.Text(
    value='2025-01-01',
    description='Dự đoán ngày:',
)

button = widgets.Button(description="Dự đoán")

output = widgets.Output()

# ===== FUNCTION VẼ + DỰ ĐOÁN =====
def update_chart(b):
    output.clear_output()
    
    stock = dropdown.value
    
    data = df[['Date', stock]].dropna()
    
    # ===== TẠO FEATURE =====
    data['Days'] = (data['Date'] - data['Date'].min()).dt.days
    
    X = data[['Days']]
    y = data[stock]
    
    # ===== TRAIN MODEL =====
    model = LinearRegression()
    model.fit(X, y)
    
    # ===== DỰ ĐOÁN =====
    try:
        future_date = pd.to_datetime(date_input.value)
        future_days = (future_date - data['Date'].min()).days
        
        pred = model.predict(pd.DataFrame([[future_days]], columns=['Days']))[0]
        
    except:
        pred = None
    
    # ===== VẼ BIỂU ĐỒ =====
    fig = go.Figure()
    
    # dữ liệu thật
    fig.add_trace(go.Scatter(
        x=data['Date'],
        y=y,
        mode='lines',
        name='Giá thật'
    ))
    
    # đường hồi quy
    fig.add_trace(go.Scatter(
        x=data['Date'],
        y=model.predict(X),
        mode='lines',
        name='Dự đoán',
        line=dict(dash='dash')
    ))
    
    # điểm dự đoán
    if pred:
        fig.add_trace(go.Scatter(
            x=[future_date],
            y=[pred],
            mode='markers+text',
            name='Điểm dự đoán',
            text=[f"{round(pred,2)}"],
            textposition="top center",
            marker=dict(size=10)
        ))
    
    # ===== STYLE =====
    fig.update_layout(
        title=f"Dashboard cổ phiếu {stock}",
        xaxis_title="Ngày",
        yaxis_title="Giá",
        template="plotly_dark",
        height=500
    )
    
    with output:
        fig.show()
        if pred:
            print(f"📈 Giá dự đoán: {round(pred,2)}")

# ===== EVENT =====
button.on_click(update_chart)

# ===== HIỂN THỊ =====
display(dropdown, date_input, button, output)
```
